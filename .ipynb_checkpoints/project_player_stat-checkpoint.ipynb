```r
library(tidyverse)
library(dbplyr)
library(DBI)
library(repr)
```

```r
player_stat = read_csv("https://drive.google.com/uc?export=download&id=1_MECmUXZuuILYeEOfonSGqodW6qVdhsS")
```

```r
# Clean data
colnames(player_stat) = gsub(" ", "_", colnames(player_stat))

# player_star: raw
# player_clean: cleaned Age, Current_Rank, Best_Rank, Seasons, Prize_Money
# player_select: selected Name, Age, Current_Rank, Best_Rank, Prize_Money
player_clean <- player_stat |>
separate(col = Age, into = c("Age", "DateOfBirth"), sep = " ", convert = TRUE) |>
separate(col = Current_Rank, into = c("Current_Rank", "Points"), sep = " ", convert = TRUE) |>
separate(col = Best_Rank, into = c("Best_Rank", "Date_Of_Best_Rank"), sep = " ", convert = TRUE)
# arrange(Current_Rank)

player_clean$Seasons <- as.numeric(player_stat$Seasons)
player_clean$Prize_Money <- gsub("\\D+", "", as.character(player_clean$Prize_Money))
player_clean$Prize_Money <- as.numeric(player_clean$Prize_Money)

player_select <- select(player_clean, Name, Age, Current_Rank, Best_Rank, Prize_Money)
```

```r
# Prize Money VS Current Rank

current_prize_plot <- ggplot(player_clean, aes(x = Current_Rank, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Current Rank", y = "Prize Money", color = "Active")

current_prize_plot

# filtered out extreme large prize money
player_filter_low_prize <- filter(player_clean, Prize_Money < 1e+08)
current_low_prize_plot <- ggplot(player_filter_low_prize, aes(x = Current_Rank, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Current Rank", y = "Prize Money", color = "Active")

current_low_prize_plot
```

```r
# Best Rank VS Prize Money

best_prize_plot <- ggplot(player_clean, aes(x = Best_Rank, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Best Rank", y = "Prize Money", color = "Active")

best_prize_plot

best_low_prize_plot <- ggplot(player_filter_low_prize, aes(x = Best_Rank, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Best Rank", y = "Prize Money", color = "Active")

best_low_prize_plot
```

```r
# Age VS Prize Money
best_prize_plot <- ggplot(player_clean, aes(x = Age, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Age", y = "Prize Money", color = "Active")

best_low_prize_plot <- ggplot(player_filter_low_prize, aes(x = Age, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Age", y = "Prize Money", color = "Active")

best_prize_plot
best_low_prize_plot
```

```r
# Seasons played VS Prize Money
season_prize_plot <- ggplot(player_clean, aes(x = Seasons, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Seasons played", y = "Prize Money", color = "Active")

season_low_prize_plot <- ggplot(player_filter_low_prize, aes(x = Seasons, y = Prize_Money, color = Active)) +
geom_point() +
labs(x = "Seasons played", y = "Prize Money", color = "Active")

season_prize_plot
season_low_prize_plot
```

```r
# Best Rank VS Current Rank
current_best_rank_plot <- ggplot(player_clean, aes(x = Best_Rank, y = Current_Rank, color = Active)) +
geom_point() +
labs(x = "Best Rank", y = "Current Rank", color = "Active")

current_best_rank_plot
```

```r
# Seasons played VS Current Rank
current_seasons_plot <- ggplot(player_clean, aes(x = Seasons, y = Current_Rank, color = Active)) +
geom_point() +
labs(x = "Seasons", y = "Current Rank", color = "Active")

current_seasons_plot
```
